## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
We frame this as a classification task. We create is_declining_label by relabeling trend_direction (down → 1, not down → 0) — this step is plain data transformation, not ML. We then train a classifier on this label using the page's other features. The classifier learns patterns connecting features of a page to its decline, and produces both a final yes/no prediction and, more usefully for us, a confidence probability for each page. We sort all pages by this probability, descending, to produce our ranked review queue.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
.We will predict a proxy label. the is_declining_label (relabelled from trend-direction, where trend direction = down is equivalent to is_declining_label = 1) is a proxy label because it is used as a stand-in for "this page needs review" which is never directly measuured in the data or labelled by a human. There might be pages that are declining but their trend direction isnt down and there might be pages that have their trend direction down due to some reason but they arent declining in reality. This is_declining_label comes from a defined rule which is that if the trend direction is down then the page is declining.
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
Precision@K where K is 50 for example. Precision@50 asks the qustion that of the top 50 pages that were predicted by the model as requiring review, what percentage actually had is_declining_label = 1 (i.e the proxy label was predicted correctly. we cant say that the percentage that actually required review because is_declining is just a proxy label for requiring review.). Notebook showed how the baseline rule gave Precision@50 to be only 0.24 while our ml approach almost tripled it to 0.74. This number can be considered good because firstly it is much greater than the precision we got using our hand written rule and secondly it shows that more than 50 percent (almost 74 percent) of the pages that the model flagged for review, were actually declining. we can not, however say for sure if the pages required review (because is declining is just a proxy) or if the review helped pages improve (because we haven't been provided any such information).
```
