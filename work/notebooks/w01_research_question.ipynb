## 3. Quick look at the data (2-3 real numbers)

*Load the starter CSV below and show 2-3 real numbers that make your lane look worth the next 7 weeks.*

```python
import pandas as pd

# Load the same starter dataset you've already used in notebooks 1 and 2
df = pd.read_csv("content_refresh_anonymized.csv")
# Just look at it — get oriented before computing anything specific
print(df.shape)        # how many rows, how many columns
df.head(10)             # see actual rows
df.describe()            # min/max/mean for every numeric column — helps spot the scale of things
df[["days_since_last_update", "impressions_90d", "avg_position", "ctr"]].describe()
# we see that 75 percent of the pages haven't been updated since 104 days so a possible baseline rule for review due to possible staleness can be: days_since_last_update > 104.
# also to filter out pages that haven't been viewed for a long time/have impressions less than the median value/50th percentile of 731, we can use the rule to only review if impressions_90d > 731
# this should be coupled with a page that also has a visibly low ctr (click-through rate) as this would mean that the page does have have views/is visible, but isnt being clicked at. to filter such pages out we use the rule: ctr < 0.07, where 0.07 is the median ctr.
# Apply the rule and see how many pages it actually flags
worth_reviewing = (
    (df["days_since_last_update"] > 104) &
    (df["impressions_90d"] > 731) &
    (df["ctr"] < 0.07)
)
count_flagged = worth_reviewing.sum()
pct_flagged = count_flagged / len(df) * 100

print(f"Pages flagged for review: {count_flagged} ({pct_flagged:.1f}% of all pages)")

# checking if the flagged pages are actually more likely to be declining
# than the average page
overall_decline_rate = df["trend_direction"].str.lower().eq("down").mean() * 100
flagged_decline_rate = df.loc[worth_reviewing, "trend_direction"].str.lower().eq("down").mean() * 100

print(f"Decline rate overall: {overall_decline_rate:.1f}%")
print(f"Decline rate among flagged pages: {flagged_decline_rate:.1f}%")
#the findings show that the decline rate among flagged pages is quite high (around 80 percent) but the number of pages flagged is quite low. This is perhaps because the conditions i set were quite constrained. I changed the 75th percentile of impressions to 25th percentile and that flagged about 0.2 percent pages but the decline rate was still around 80. Hoping to explore this further in the coming weeks.
```
